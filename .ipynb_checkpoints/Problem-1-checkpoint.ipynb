```python
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
n_steps = np.arange(501)
place = np.array([0])
prob = np.random.random((500,))
prob = prob - 0.5

for i in range(500):
    place = np.append(place, place[i] + prob[i])

plt.plot(n_steps, place)
```

```python
def total(steps, n):
    out = np.zeros((n,))
    for i in range(n):
        prob = np.random.random((steps,))
        prob = prob - 0.5
        final_position = np.sum(prob)
        out[i] = final_position
    return out

out1 = total(500, 1000)
plt.hist(out1, bins=100)
plt.show()
```

```python
import pylab
import random 

n = 10000

x = np.zeros(n) 

for i in range(1, n): 
	val = random.randint(1, 2) 
	if val == 1: 
		x[i] = x[i - 1] + 1
	elif val == 2: 
		x[i] = x[i - 1] - 1


pylab.plot(x) 
pylab.show() 
```

```python
def discrete(steps, n):
    out = np.zeros((n,))
    for i in range(n):
        prob = np.random.randint(2, size=steps)
        prob = prob * 0.5
        final_position = np.sum(prob)
        out[i] = final_position
    return out
```

```python
out2 = discrete(500, 1000)
plt.hist(out2, bins=100)
plt.show()
```
