# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
def One_dim(n_step):
    step_x = np.random.random(n_step)-1/2
    
    return (np.cumsum(step_x))

n_step = 1
n_walk = 100

x_f = np.zeros(n_walk)
for run in range(n_walk):
    x = One_dim(n_step)
    x_f[run] = x[-1]
    
pylab.plot(x_f,) 
pylab.show() 
```

```python
arr = np.array(x_f)

sub_arr = np.abs(arr[:,None] - arr)


pylab.hist(sub_arr[0],bins=100) 
pylab.show() 
```

```python
import pylab
import random 

n = 10000

x = np.zeros(n) 

for i in range(1, n): 
	val = random.randint(1, 2) 
	if val == 1: 
		x[i] = x[i - 1] + 1
	elif val == 2: 
		x[i] = x[i - 1] - 1


pylab.plot(x) 
pylab.show() 
```

```python
arr = np.array(x)

sub_arr = np.abs(arr[:,None] - arr)


pylab.hist(sub_arr[0],bins=100) 
pylab.show() 
```
