# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
def path(walk):
    location_x = np.array([0])
    location_y = np.array([0])
    
    for i in range(len(walk[0])):
        location_x = np.append(location_x, location_x[i] + walk[0][i])
        location_y = np.append(location_y, location_y[i] + walk[1][i])

    return [location_x, location_y]
```

```python
prob_x = np.random.random((500,)) - 0.5
prob_y = np.random.random((500,)) - 0.5

location = path([prob_x, prob_y])

plt.plot(location[0], location[1])
```

```python
plt.figure(figsize=(25, 6))

r = np.sqrt(location[1]**2 + location[1]**2)

plt.subplot(131)
pylab.plot(location[0]) 


plt.subplot(132)
pylab.plot(location[1]) 


plt.subplot(133)
pylab.plot(r) 
pylab.show() 
```

```python
def r(location):
    out = location[0] * location[1]
    return out

def continue_case(steps, n):
    out_x = np.zeros((n,))
    out_y = np.zeros((n,))
    for i in range(n):
        prob_x = np.random.random((steps,)) - 0.5
        prob_y = np.random.random((steps,)) - 0.5
        final_location_x = np.sum(prob_x)
        final_location_y = np.sum(prob_y)
        out_x[i] = final_location_x
        out_y[i] = final_location_y
    return [out_x, out_y]


def ran_step_2d_1(n_steps, step_size, n):
    out_x = np.zeros((n,))
    out_y = np.zeros((n,))
    for i in range(n):
        prob_x = np.random.randint(2, size=n_steps)
        prob_y = np.random.randint(2, size=n_steps)
        prob_x = step_size * (prob_x - 0.5)
        prob_y = step_size * (prob_y - 0.5)
        out_x[i] = np.sum(prob_x)
        out_y[i] = np.sum(prob_y)
    out = np.array([out_x, out_y])
    return out
```

```python
out1 = ran_step_2d_1(500, 0.5, 1000)
plt.hist(r(out1), bins=50)
plt.show()
```

```python
def r(location):
    out = location[0] * location[1]
    return out

def rand_ang(steps, n):
    out_x = np.zeros((n,))
    out_y = np.zeros((n,))

    for i in range(n):
        prob = np.random.random((steps,))
        prob = prob * 2 * np.pi
        x = np.cos(prob)
        y = np.sin(prob)
        out_x[i] = np.sum(x)
        out_y[i] = np.sum(y)

    out = np.array([out_x, out_y])
    return out
```

```python
out3 = rand_ang(500, 1000)

plt.hist(r(out3), bins=100)
plt.show()
```
